```python
import cv2
import numpy as np
import os
import random

def create_word_image(word, output_filename="word_image.png", width=200, height=100, font_scale=1, font_thickness=2):
    """生成包含指定单词的白底黑字图片。"""
    img = np.ones((height, width, 3), np.uint8) * 255
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_color = (0, 0, 0)
    (text_width, text_height), baseline = cv2.getTextSize(word, font, font_scale, font_thickness)
    x = (width - text_width) // 2
    y = (height + text_height) // 2
    cv2.putText(img, word, (x, y), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
    cv2.imwrite(output_filename, img)
    print(f"生成了图像: {output_filename}")


def add_random_line_erasures(image_path, output_path, num_lines=15, line_length=50, line_thickness=3, eraser_color=(255, 255, 255)):
    """在图像上添加随机线段擦除效果。"""
    img = cv2.imread(image_path)
    if img is None:
        print(f"无法加载图像: {image_path}")
        return

    height, width, _ = img.shape

    for _ in range(num_lines):
        # 随机选择线段的起始点
        x1 = random.randint(0, width - 1)
        y1 = random.randint(0, height - 1)

        # 随机选择线段的角度 (0-360 度)
        angle = random.uniform(0, 360)

        # 计算线段的终点
        x2 = int(x1 + line_length * np.cos(np.radians(angle)))
        y2 = int(y1 + line_length * np.sin(np.radians(angle)))

        # 确保线段终点在图像范围内
        x2 = max(0, min(x2, width - 1))
        y2 = max(0, min(y2, height - 1))

        # 绘制线段
        cv2.line(img, (x1, y1), (x2, y2), eraser_color, line_thickness)

    cv2.imwrite(output_path, img)
    print(f"添加了线段擦除效果，并保存到: {output_path}")


def process_words_from_file(input_file="1.txt", output_dir="word_images", erased_output_dir="erased_word_images"):
    """从文件中读取单词，生成图像，并添加随机擦除效果。"""
    try:
        with open(input_file, "r", encoding="utf-8") as f:
            words = [line.strip() for line in f]
    except FileNotFoundError:
        print(f"错误: 文件 '{input_file}' 未找到.")
        return

    # 创建输出目录（如果不存在）
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    if not os.path.exists(erased_output_dir):
        os.makedirs(erased_output_dir)

    for i, word in enumerate(words):
        # 生成原始图像
        original_image_path = os.path.join(output_dir, f"{i+1}.png")
        create_word_image(word, original_image_path)

        # 添加擦除效果并保存
        erased_image_path = os.path.join(erased_output_dir, f"{i+1}.png")
        add_random_line_erasures(original_image_path, erased_image_path)


if __name__ == "__main__":
    process_words_from_file()
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

with open('1.txt', 'w', encoding='utf-8') as f:  # 以写入模式打开文件，并指定编码为utf-8
    for i in range(df.shape[0]):
        try:
            str_val = (df.iloc[i, 0])  # 确保读取的是字符串，避免可能的类型错误
        except TypeError as e:
            print(f"Error converting value to string at row {i}: {e}")
            continue  # 跳过当前行

        if len(str_val) < 5:
            continue
        if ' ' in str_val:
            continue
        f.write(str_val + '\n')  # 将字符串写入文件，并添加换行符
        print(str_val)  # 仍然打印到控制台，方便查看进度

print("数据已写入到 1.txt 文件中。")
```
