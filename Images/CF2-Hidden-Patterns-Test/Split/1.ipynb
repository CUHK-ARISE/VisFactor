```python
from PIL import Image
import os

# 确保输出目录存在
output_dir = "processed_images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 处理50张图片
for i in range(1, 51):
    input_filename = f"{i}.jpg"
    
    # 检查原始图片是否存在
    if not os.path.exists(input_filename):
        print(f"警告: 文件 {input_filename} 不存在，跳过处理")
        continue
    
    try:
        # 打开原始图片
        img = Image.open(input_filename)
        
        # 1. 原始图片 (旋转0度)
        output_filename = f"{i}.jpg"
        img.save(os.path.join(output_dir, output_filename))
        
        # 2. 旋转90度
        output_filename = f"{i+50}.jpg"
        img_90 = img.rotate(-90, expand=True)  # 逆时针旋转90度
        img_90.save(os.path.join(output_dir, output_filename))
        
        # 3. 旋转180度
        output_filename = f"{i+100}.jpg"
        img_180 = img.rotate(180, expand=True)
        img_180.save(os.path.join(output_dir, output_filename))
        
        # 4. 旋转270度
        output_filename = f"{i+150}.jpg"
        img_270 = img.rotate(-270, expand=True)  # 逆时针旋转270度
        img_270.save(os.path.join(output_dir, output_filename))
        
        # 5. 上下翻转
        output_filename = f"{i+200}.jpg"
        img_flip_tb = img.transpose(Image.FLIP_TOP_BOTTOM)
        img_flip_tb.save(os.path.join(output_dir, output_filename))
        
        # 6. 左右翻转
        output_filename = f"{i+250}.jpg"
        img_flip_lr = img.transpose(Image.FLIP_LEFT_RIGHT)
        img_flip_lr.save(os.path.join(output_dir, output_filename))
        
        print(f"已处理图片 {input_filename}")
        
    except Exception as e:
        print(f"处理图片 {input_filename} 时出错: {e}")
        
print("处理完成！所有图片已保存到", output_dir)
```
